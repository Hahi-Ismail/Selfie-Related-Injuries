Selfie Related Injuries


about this dataset

This dataset contains a range of cases of serious injury or even death as a result of taking a selfie, indicating the countries and age groups in which these cases of injury or death occurred.

Data preparation and processing

#descriptive statistics

````python
df.describe()
````

It is clear from the descriptive statistics that the lowest number of injuries is one and the highest number is 48

A value of 48 may be an outlier; So we will arrange the data according to the number of infections to make sure of that

````python
#Count the values in Injuries/Casualties column 
df.value_counts("Injuries/Casualties")
````

We can see that the largest of the values is 48 and that the value immediately before it is 11
We can consider that 48 is an odd value, but by research it was found that the number is correct and was not the result of an error in data entry or otherwise, as it was found that 48 people were injured when standing too close to a burning bakery in Chennai, India, to take selfies. They repeatedly ignored warnings advising them to move away from the blaze

Now we will look at the number of infections in each country

````python
plt.figure(figsize=(30,20))
sns.countplot(x= "Country", data=df, order=df['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=25)
plt.tick_params(axis='y', which='major', labelsize=25)
plt.title("Distribution of Countries", fontsize=40)
plt.xlabel("Country", fontsize=40)
plt.ylabel("Count", fontsize=40)
plt.show()
````

It turns out that India is the most frequent country in the data set, followed by the United States, Russia and Pakistan, then we can classify the rest of the countries into five groups with equal frequency.

Now we will look at the number of infections in each country

````python
plt.figure(figsize=(20,15))
sns.countplot(x= "Type", data=df, order=df['Type'].value_counts().index)
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=25)
plt.tick_params(axis='y', which='major', labelsize=25)
plt.title("Distribution of Types", fontsize=40)
plt.xlabel("Type", fontsize=40)
plt.ylabel("Count", fontsize=40)
plt.show()
````

It is clear from the previous figure the arrangement of the types of injuries in terms of the number of recurrences, in order from least to most

````python
df.groupby(['Country'])['Injuries/Casualties'].sum().sort_values(ascending=False)
````

````python
df2 = df.groupby(['Country'])['Injuries/Casualties'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
sns.barplot(x="Country", y="Injuries/Casualties", data=df2, palette="Set2")
plt.title("Selfie-Related Injuries by Country")
plt.xlabel("Country", labelpad=20)
plt.ylabel("Number of Injuries", labelpad=20)
plt.xticks(rotation=90)
plt.show()
````

The previous figure shows the number of injuries in each country, arranged from highest to lowest, without clarifying the types of injuries

````python
df2 = df.groupby(['Type'])['Injuries/Casualties'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
sns.barplot(x="Type", y="Injuries/Casualties", data=df2, palette="Set2")
plt.title("Selfie-Related Injuries by Type")
plt.xlabel("Type", labelpad=20)
plt.ylabel("Number of Injuries", labelpad=20)
plt.xticks(rotation=90)
plt.show()
````

From the previous figure, it is clear that drowning, falling, and transportation, respectively, are the most common causes of accidents related to taking selfies

````python
PT = df.pivot_table(columns="Type", index="Country", values="Injuries/Casualties", aggfunc='sum')
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
sns.heatmap(PT, cmap="coolwarm")
plt.title("Selfie-Related Injuries by Country and Type")
plt.xlabel("Type", labelpad=20)
plt.ylabel("Country", labelpad=20)
plt.show()
````

We can, through the previous chart, distribute the types and numbers of injuries according to the country. It is clear that injuries due to falls are the highest in terms of geographical spread, while injuries due to fire did not occur only once in India, but caused 48 people to be injured.
